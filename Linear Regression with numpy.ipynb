```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# area of a house and its price
area = [1.9,2.4,5.3,3.6,2.5,4.2,5.2,6.3,1.2,3.6,4.3,4.1,6.2,3.8,4.5]
price= [210,250,480,390,220,460,490,580,180,350,410,450,580,450,425]
```

```python
x=np.array(area)
y=np.array(price)
plt.figure(figsize=(12,5))
plt.scatter(x,y,marker='o',c='Red')
plt.show()
```

```python
#lets take initally the line at slope of 2.5 and intercept at 10.9
m=2.5
c=10.9
LR=0.003 #learning Rate

for i in range(500):
    yhat =m*x + c
    
    #applying gradient decent algorithm to decrease the value
    dedm=2*np.dot((yhat-y),x)
    m=m -LR *dedm
    
    dedc=2*np.dot((yhat-y),np.ones(y.size))
    c=c -LR *dedc
    
    error = np.mean((y-yhat)**2)
    print(error)
    
```

```python
print(m,c)
```

# Linear Regression with Sklearn
