# Data Visualization


matplotlib.pyplot as plt

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[5,10,15,20,25,30,35,40,45]
```

# line plot

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y,marker='o')
plt.show()
```

 Marker | line | color

```python
plt.plot(y,'*:r')
```

```python
#ms-markersize
#mec=markeredgecolor
#mfc=markerfacecolor
plt.plot(y,'*-m',ms=10,mec='r',mfc='y',linewidth=2.5)
plt.show()
```

# Scatter plot(dot dot dot)

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,marker='+',c='r',edgecolor='g',s=50)
plt.show()
```

# Histogram 

it is used to display the distribution of a continuous attribute,such as salary ,age,temperature 

```python
temp=[25,26,28,24,26,26,28,28,25,23,21,24,26,26,24,25]
plt.hist(temp)
plt.xlabel('temperature')
plt.ylabel('count')
plt.show()
```

```python
temp=[25,26,28,24,26,26,28,28,25,23,21,24,26,26,24,25]
plt.hist(temp,bins=3)
#bins -> is used to set how many bins do you need
plt.xlabel('temperature')
plt.ylabel('count')
plt.show()
```

```python
temp1=[25,26,28,24,26,26,28,28,25,23,21,24,26,26,24,25]
temp2=[23,23,23,25,25,25,25,24,24,24,25,25,21,21,21,21]
plt.hist(temp1)
plt.hist(temp2,alpha=0.5)
#alpha-> brightness 
plt.show()
```

# Pie chart

    it is also used for comparison 

```python
values=[12,50,32,26,45]
plt.pie(values)
plt.show()
```

```python
values=[12,50,32,26,45]
names=['rahul','susith','nandha','vimal','manoj']
plt.pie(values,labels=names,autopct='%.f',explode=[0.3,0,0,0,0])
plt.show()
```

# Boxplot

in boxplot we can find maximum minimum mean and as well as outliers

```python
temp1=[25,26,28,24,26,26,28,28,25,23,21,24,26,26,24,25]
temp1.append(35)
temp1.append(39)
temp1
```

```python
plt.boxplot(temp1)
plt.ylim([20,50])
plt.grid(True)
plt.show()
```
